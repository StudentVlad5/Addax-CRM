Трансляція тензорів (broadcasting)

Це спосіб, завдяки якому в бібліотеці PyTorch  можна виконувати операції між тензорами різних розмірів. Головне, щоб ці розміри були "сумісними", тобто відповідали певним правилам.

Уявіть, що трансляція — це як "розтягування" меншого тензора до розмірів більшого, щоб можна було застосувати операцію між ними, наприклад, додавання чи множення. Це дуже зручно, особливо при роботі з великими обсягами даних у глибокому навчанні, де потрібні операції над групами даних одночасно.

Основні правила трансляції

1. Необхідність хоча б одного виміру: Тензори повинні мати хоча б одну розмірність. Тобто не можна виконати трансляцію з повністю порожніми тензорами.

2. Сумісність вимірів: Перевірка йде з кінця розмірів до початку. Порівнюються останні виміри обох тензорів:

3. Один із розмірів дорівнює 1 — тоді цей розмір буде "розтягнуто" до більшого розміру.

4. Обидва розміри однакові — тоді вони сумісні.
Один із тензорів не має виміру — тоді цей тензор "розширюється" для відповідності іншому тензору.
5. Якщо виміри не відповідають жодному з цих правил, трансляція не можлива, і PyTorch видасть помилку.

ПРИКЛАД

```python
a = torch.ones(4, 3, 2)        # Розміри (4, 3, 2)

# Тензор b має розміри (3, 2), тому PyTorch автоматично "розширить" його до (4, 3, 2)
b = a * torch.rand(3, 2)        # Результат буде розміром (4, 3, 2)
b.size()
```

```python
# Тензор c має розміри (3, 1), тому PyTorch "розтягне" останній розмір до 2, щоб він збігався з (4, 3, 2)
c = a * torch.rand(3, 1)        # Результат буде розміром (4, 3, 2)
c.size()
```

```python
# Тензор d має розміри (1, 2), PyTorch "розтягне" перший розмір до 3, а другий залишиться 2
d = a * torch.rand(1, 2)        # Результат також буде розміром (4, 3, 2)
d.size()
```

Опис датасету сортів родзинок "Kecimen" і "Besni"
Цей датасет містить морфологічні характеристики двох сортів родзинок — Kecimen та Besni. Він складається з даних про 900 родзинок, розділених на 450 зразків для кожного сорту. Зображення родзинок були отримані за допомогою системи комп'ютерного зору (CVS), і для кожної родзинки було витягнуто сім морфологічних ознак. Основною метою цього датасету є аналіз та класифікація сортів родзинок на основі їх геометричних параметрів.

**Характеристики даних:**
**Area (Площа):**

Опис: Це кількість пікселів всередині меж родзинки. Цей показник вказує на розмір родзинки.
Діапазон значень: від 25,387 до 235,047 пікселів.

**Major Axis Length (Довжина головної осі):**

Опис: Це довжина найдовшої лінії, яку можна провести через родзинку.
Діапазон значень: від 225.63 до 997.29 одиниць довжини.

**Minor Axis Length (Довжина малої осі):**

Опис: Це довжина найкоротшої лінії, яку можна провести через родзинку, перпендикулярно до головної осі.
Діапазон значень: від 143.71 до 492.28 одиниць довжини.

**Eccentricity (Ексцентриситет):**

Опис: Показник того, наскільки форма родзинки відхиляється від ідеальної окружності. Це вимірює еліптичність форми родзинки, де ексцентриситет наближається до 0 для ідеальних круглих форм, а значення ближчі до 1 означають більш витягнуту форму.
Діапазон значень: від 0.35 до 0.96.

**Convex Area (Площа опуклої оболонки):**

Опис: Це площа найменшої опуклої оболонки, яка може вмістити родзинку. Це важливий параметр для оцінки форми та компактності об'єкта.
Діапазон значень: від 26,139 до 278,217 пікселів.

**Extent (Простягання):**

Опис: Це відношення площі родзинки до площі її обмежувального прямокутника. Простягання показує, наскільки добре родзинка займає простір в межах своєї обмежувальної рамки.
Діапазон значень: від 0.38 до 0.84.

**Perimeter (Периметр):**

Опис: Це периметр родзинки, виміряний як відстань між межами родзинки і пікселями навколо неї. Цей параметр відображає складність форми об'єкта.
Діапазон значень: від 619.07 до 2697.75 одиниць довжини.

```python
X = df.drop(columns='Class').copy()
y = df['Class']

# Кодуємо категоріальні ознаки
cat_features = X.select_dtypes('object').columns
for col in cat_features:
    X[col], _ = X[col].factorize()
```

```python
cat_features = X.select_dtypes('object').columns
for col in cat_features:
    X[col], _ = X[col].factorize()

# Обчислення MI для класифікації
mi_scores = mutual_info_classif(
    X, y,
    discrete_features=X.columns.isin(cat_features.to_list()),
    random_state=42
)

# Створення Series
mi_scores = pd.Series(mi_scores, name='MI Scores', index=X.columns).sort_values()

mi_scores.sample(6)
```

```python
# Відбір ознак: MI > 0
useful_features = mi_scores[mi_scores > 0].index.tolist()

# Відкидаємо надмірно корельовані пари (|r| > 0.95)
corr_matrix = X[useful_features].corr().abs()
upper = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(bool))

to_drop = set()
for col in upper.columns:
    for row in upper.index:
        if upper.loc[row, col] > 0.95:
            # порівняємо MI: залишимо з більшою важливістю
            if mi_scores[row] < mi_scores[col]:
                to_drop.add(row)
            else:
                to_drop.add(col)

final_features = [f for f in useful_features if f not in to_drop]

# Підсумковий датасет
X_selected = X[final_features]

print("\nВідібрані ознаки для моделі:")
print(final_features)

# Якщо потрібно — об’єднати з цільовою змінною:
df_selected = pd.concat([X_selected, y], axis=1)
```

```python
# Демонстрація лінійного шару
m = nn.Linear(4, 1)  # Лінійний шар з 4 входами та 1 вихід
input = torch.randn(4, 4)  # Випадковий вхідний тензор розміром (4, 4)
output = m(input)  # Обчислення виходу

print('Input:', input, f'shape {input.shape}', sep='\n')
print('\nOutput:', output, f'shape {output.shape}', sep='\n')
```

```python
# Define the logistic regression model

class LogisticRegression(nn.Module):
    def __init__(self, input_dim):
        super(LogisticRegression, self).__init__()
        self.linear = nn.Linear(input_dim, 1)  # Один вихідний нейрон для бінарної класифікації
        self.sigmoid = nn.Sigmoid()

    def forward(self, x):
        out = self.linear(x)
        out = self.sigmoid(out)
        return out
```

```python
# Instantiate the model
input_dim = X_train.shape[1]
model = LogisticRegression(input_dim)

# Define loss function and optimizer
criterion = nn.BCELoss()
optimizer = optim.SGD(model.parameters(), lr=0.01)
```

```python
model.parameters
```

```python
# Навчання моделі
num_epochs = 500
for epoch in range(num_epochs):
    # Прямий прохід
    outputs = model(X_train)
    loss = criterion(outputs, y_train)

    # Зворотний прохід і оптимізація
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()

    if (epoch + 1) % 5 == 0:
        print(f'Epoch [{epoch + 1}/{num_epochs}], Loss: {loss.item():.4f}')
```

```python
# Тестування моделі
with torch.no_grad():
    y_pred = model(X_test).squeeze().numpy()
    y_pred = (y_pred > 0.5).astype(int)  # Використання порогу 0.5
    accuracy = accuracy_score(y_test.numpy(), y_pred)
    print('Accuracy:', accuracy)
```

```python
#  Створення моделі нейронної мережі
class NObeyesdadNN(nn.Module):
    def __init__(self, input_size, num_clas):
        super(NObeyesdadNN, self).__init__()

        self.fc1 = nn.Linear(input_size, 256)   # Збільшено кількість нейронів у першому шарі
        self.bn1 = nn.BatchNorm1d(256)           # Batch Normalization для 1-го шару
        self.fc2 = nn.Linear(256, 128)           # Збільшено кількість нейронів у другому шарі
        self.bn2 = nn.BatchNorm1d(128)           # Batch Normalization для 2-го шару
        self.fc3 = nn.Linear(128, num_clas)      # Вихідний шар
        self.relu = nn.ReLU()                    # Активація ReLU для схованих шарів
        self.dropout = nn.Dropout(0.5)           # Dropout для регуляризації

    def forward(self, x):
        x = self.relu(self.bn1(self.fc1(x)))  # Перший схований шар з Batch Normalization
        x = self.dropout(x)                    # Dropout після першого шару
        x = self.relu(self.bn2(self.fc2(x)))  # Другий схований шар з Batch Normalization
        x = self.dropout(x)                    # Dropout після другого шару
        x = self.fc3(x)                        # Вихідний шар
        return x
```

```python
# Датасет
class HouseDataset(Dataset):
    def __init__(self, X, y):
        self.X = torch.tensor(X, dtype=torch.float32)
        self.y = torch.tensor(y, dtype=torch.float32)

    def __len__(self):
        return len(self.y)

    def __getitem__(self, idx):
        return self.X[idx], self.y[idx]
```

```python
train_dataset = HouseDataset(X_train, y_train)
test_dataset = HouseDataset(X_test, y_test)
```

```python
# Loading Batches of Data

train_dataloader = DataLoader(train_dataset,
                              batch_size=200,
                              num_workers=4
                             )

test_dataloader = DataLoader(test_dataset,
                              batch_size=200,
                              num_workers=4
                             )
```

```python
import torch.nn as nn
class LinearModel(nn.Module):
    def __init__(self, in_dim, out_dim=1):
        super().__init__()

        self.features = nn.Sequential(
            nn.Linear(in_dim, 256),
            nn.BatchNorm1d(256),
            nn.ReLU(),
            nn.Dropout(0.3),

            nn.Linear(256, 128),
            nn.BatchNorm1d(128),
            nn.ReLU(),
            nn.Dropout(0.3),

            nn.Linear(128, 64),
            nn.BatchNorm1d(64),
            nn.ReLU(),
            nn.Dropout(0.3),

            nn.Linear(64, out_dim)
        )

    def forward(self, x):
        return self.features(x)
```

```python
# Initialize the model
model = LinearModel(in_dim=X_train.shape[1], out_dim=1)

# Define the loss function and optimizer
criterion = nn.MSELoss()
optimizer = torch.optim.Adam(model.parameters(), lr=1e-4)

train_losses, train_rmses = [], []
test_losses, test_rmses= [], []
```

```python
num_epochs = 200

# Функція RMSE на основі MSELoss
def RMSE(y_true, y_pred):
    mse_loss = torch.nn.MSELoss()
    y_true = torch.tensor(y_true, dtype=torch.float32)
    y_pred = torch.tensor(y_pred, dtype=torch.float32)
    mse = mse_loss(y_pred, y_true)
    return torch.sqrt(mse)

for epoch in range(num_epochs):

    # Train step
    model.train()

    y_pred_train = []
    y_true_train = []

    for data in train_dataloader:
        # Отримуємо та готуємо дані
        inputs, targets = data
        inputs, targets = inputs.float(), targets.float()
        targets = targets.reshape((targets.shape[0], 1))

        # Прогноз (Forward pass)
        outputs = model(inputs)
        loss = criterion(outputs, targets)

        # Оновлення ваг
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

        # Зберігаємо прогнози і справжні значення
        y_pred_train.extend(outputs.cpu().detach().numpy())
        y_true_train.extend(targets.cpu().detach().numpy())

    # Обчислюємо RMSE кожні 10 епох
    if (epoch+1) % 10 == 0:
        rmse_train = RMSE(y_true_train, y_pred_train)
        print(f'Epoch [{epoch+1}/{num_epochs}], Loss: {loss.item():.4f}, RMSE: {rmse_train:.4f}')

    train_rmses.append(RMSE(y_true_train, y_pred_train))
    train_losses.append(loss.cpu().detach().numpy())

    # Eval step
    model.eval()

    y_pred_test = []
    y_true_test = []

    with torch.no_grad():
        for data in test_dataloader:
            # Отримуємо та готуємо дані
            inputs, targets = data
            inputs, targets = inputs.float(), targets.float()
            targets = targets.reshape((targets.shape[0], 1))

            # Прогноз (Forward pass)
            outputs = model(inputs)
            loss = criterion(outputs, targets)

            # Зберігаємо прогнози і справжні значення
            y_pred_test.extend(outputs.cpu().detach().numpy())
            y_true_test.extend(targets.cpu().detach().numpy())

        # Обчислюємо RMSE для тестових даних
        test_rmses.append(RMSE(y_true_test, y_pred_test))
        test_losses.append(loss.cpu().detach().numpy())
```
