```python
cat_features = X.select_dtypes('object').columns
for col in cat_features:
    X[col], _ = X[col].factorize()

# –û–±—á–∏—Å–ª–µ–Ω–Ω—è MI –¥–ª—è –∫–ª–∞—Å–∏—Ñ—ñ–∫–∞—Ü—ñ—ó
mi_scores = mutual_info_classif(
    X, y,
    discrete_features=X.columns.isin(cat_features.to_list()),
    random_state=42
)

# –°—Ç–≤–æ—Ä–µ–Ω–Ω—è Series
mi_scores = pd.Series(mi_scores, name='MI Scores', index=X.columns).sort_values()

mi_scores.sample(6)
```

```python
# –í—ñ–¥–±—ñ—Ä –æ–∑–Ω–∞–∫: MI > 0
useful_features = mi_scores[mi_scores > 0].index.tolist()

# –í—ñ–¥–∫–∏–¥–∞—î–º–æ –Ω–∞–¥–º—ñ—Ä–Ω–æ –∫–æ—Ä–µ–ª—å–æ–≤–∞–Ω—ñ –ø–∞—Ä–∏ (|r| > 0.95)
corr_matrix = X[useful_features].corr().abs()
upper = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(bool))

to_drop = set()
for col in upper.columns:
    for row in upper.index:
        if upper.loc[row, col] > 0.95:
            # –ø–æ—Ä—ñ–≤–Ω—è—î–º–æ MI: –∑–∞–ª–∏—à–∏–º–æ –∑ –±—ñ–ª—å—à–æ—é –≤–∞–∂–ª–∏–≤—ñ—Å—Ç—é
            if mi_scores[row] < mi_scores[col]:
                to_drop.add(row)
            else:
                to_drop.add(col)

final_features = [f for f in useful_features if f not in to_drop]

# –ü—ñ–¥—Å—É–º–∫–æ–≤–∏–π –¥–∞—Ç–∞—Å–µ—Ç
X_selected = X[final_features]

print("\n–í—ñ–¥—ñ–±—Ä–∞–Ω—ñ –æ–∑–Ω–∞–∫–∏ –¥–ª—è –º–æ–¥–µ–ª—ñ:")
print(final_features)

# –Ø–∫—â–æ –ø–æ—Ç—Ä—ñ–±–Ω–æ ‚Äî –æ–±‚Äô—î–¥–Ω–∞—Ç–∏ –∑ —Ü—ñ–ª—å–æ–≤–æ—é –∑–º—ñ–Ω–Ω–æ—é:
df_selected = pd.concat([X_selected, y], axis=1)
```

```python
# Convert numpy arrays to PyTorch tensors
X_train = torch.tensor(X_train, dtype=torch.float32)
y_train = torch.tensor(y_train, dtype=torch.float32).unsqueeze(1)  # –î–æ–¥–∞—Ç–∏ –≤–∏–º—ñ—Ä –¥–ª—è —Å—É–º—ñ—Å–Ω–æ—Å—Ç—ñ –∑ BCELoss

X_test = torch.tensor(X_test, dtype=torch.float32)
y_test = torch.tensor(y_test, dtype=torch.float32).unsqueeze(1)  # –î–æ–¥–∞—Ç–∏ –≤–∏–º—ñ—Ä –¥–ª—è —Å—É–º—ñ—Å–Ω–æ—Å—Ç—ñ –∑ BCELoss
```

```python
# –î–µ–º–æ–Ω—Å—Ç—Ä–∞—Ü—ñ—è –ª—ñ–Ω—ñ–π–Ω–æ–≥–æ —à–∞—Ä—É
m = nn.Linear(4, 1)  # –õ—ñ–Ω—ñ–π–Ω–∏–π —à–∞—Ä –∑ 4 –≤—Ö–æ–¥–∞–º–∏ —Ç–∞ 1 –≤–∏—Ö—ñ–¥
input = torch.randn(4, 4)  # –í–∏–ø–∞–¥–∫–æ–≤–∏–π –≤—Ö—ñ–¥–Ω–∏–π —Ç–µ–Ω–∑–æ—Ä —Ä–æ–∑–º—ñ—Ä–æ–º (4, 4)
output = m(input)  # –û–±—á–∏—Å–ª–µ–Ω–Ω—è –≤–∏—Ö–æ–¥—É

print('Input:', input, f'shape {input.shape}', sep='\n')
print('\nOutput:', output, f'shape {output.shape}', sep='\n')
```

```python
# Define the logistic regression model

class LogisticRegression(nn.Module):
    def __init__(self, input_dim):
        super(LogisticRegression, self).__init__()
        self.linear = nn.Linear(input_dim, 1)  # –û–¥–∏–Ω –≤–∏—Ö—ñ–¥–Ω–∏–π –Ω–µ–π—Ä–æ–Ω –¥–ª—è –±—ñ–Ω–∞—Ä–Ω–æ—ó –∫–ª–∞—Å–∏—Ñ—ñ–∫–∞—Ü—ñ—ó
        self.sigmoid = nn.Sigmoid()

    def forward(self, x):
        out = self.linear(x)
        out = self.sigmoid(out)
        return out
```

```python
# Instantiate the model
input_dim = X_train.shape[1]
model = LogisticRegression(input_dim)

# Define loss function and optimizer
criterion = nn.BCELoss()
optimizer = optim.SGD(model.parameters(), lr=0.01)
```

```python
# –î–µ–º–æ–Ω—Å—Ç—Ä–∞—Ü—ñ—è –ª—ñ–Ω—ñ–π–Ω–æ–≥–æ —à–∞—Ä—É
m = nn.Linear(4, 1)  # –õ—ñ–Ω—ñ–π–Ω–∏–π —à–∞—Ä –∑ 3 –≤—Ö–æ–¥–∞–º–∏ —Ç–∞ 1 –≤–∏—Ö—ñ–¥
input = torch.randn(4, 4)  # –í–∏–ø–∞–¥–∫–æ–≤–∏–π –≤—Ö—ñ–¥–Ω–∏–π —Ç–µ–Ω–∑–æ—Ä —Ä–æ–∑–º—ñ—Ä–æ–º (4, 7)
output = m(input)  # –û–±—á–∏—Å–ª–µ–Ω–Ω—è –≤–∏—Ö–æ–¥—É

print('Input:', input, f'shape {input.shape}', sep='\n')
print('\nOutput:', output, f'shape {output.shape}', sep='\n')
```

```python
# Define the logistic regression model

class LogisticRegression(nn.Module):
    def __init__(self, input_dim):
        super(LogisticRegression, self).__init__()
        self.linear = nn.Linear(input_dim, 1)  # –û–¥–∏–Ω –≤–∏—Ö—ñ–¥–Ω–∏–π –Ω–µ–π—Ä–æ–Ω –¥–ª—è –±—ñ–Ω–∞—Ä–Ω–æ—ó –∫–ª–∞—Å–∏—Ñ—ñ–∫–∞—Ü—ñ—ó
        self.sigmoid = nn.Sigmoid()

    def forward(self, x):
        out = self.linear(x)
        out = self.sigmoid(out)
        return out
```

```python
# Instantiate the model
input_dim = X_train.shape[1]
model = LogisticRegression(input_dim)

# Define loss function and optimizer
criterion = nn.BCELoss()
optimizer = optim.SGD(model.parameters(), lr=0.01)
```

```python
# –ù–∞–≤—á–∞–Ω–Ω—è –º–æ–¥–µ–ª—ñ
num_epochs = 500
for epoch in range(num_epochs):
    # –ü—Ä—è–º–∏–π –ø—Ä–æ—Ö—ñ–¥
    outputs = model(X_train)
    loss = criterion(outputs, y_train)

    # –ó–≤–æ—Ä–æ—Ç–Ω–∏–π –ø—Ä–æ—Ö—ñ–¥ —ñ –æ–ø—Ç–∏–º—ñ–∑–∞—Ü—ñ—è
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()

    if (epoch + 1) % 5 == 0:
        print(f'Epoch [{epoch + 1}/{num_epochs}], Loss: {loss.item():.4f}')
```

```python
label_encoder = LabelEncoder()

# –ü–µ—Ä–µ—Ç–≤–æ—Ä—é—î–º–æ –∫–∞—Ç–µ–≥–æ—Ä—ñ–∞–ª—å–Ω—ñ –º—ñ—Ç–∫–∏ –Ω–∞ —á–∏—Å–ª–æ–≤—ñ
y_train_encoded = label_encoder.fit_transform(y_train)
y_test_encoded = label_encoder.transform(y_test)

# –ü–µ—Ä–µ—Ç–≤–æ—Ä–µ–Ω–Ω—è —Ç—Ä–µ–Ω—É–≤–∞–ª—å–Ω–∏—Ö –¥–∞–Ω–∏—Ö —É —Ç–µ–Ω–∑–æ—Ä–∏
X_train = torch.tensor(X_train.values).float()  # –ü–µ—Ä–µ—Ç–≤–æ—Ä—é—î–º–æ –Ω–∞ numpy, –∞ –ø–æ—Ç—ñ–º –Ω–∞ —Ç–µ–Ω–∑–æ—Ä —Ç–∏–ø—É float
y_train = torch.tensor(y_train_encoded).long()  # –ü–µ—Ä–µ—Ç–≤–æ—Ä—é—î–º–æ —á–∏—Å–ª–æ–≤—ñ –º—ñ—Ç–∫–∏ –Ω–∞ —Ç–µ–Ω–∑–æ—Ä —Ç–∏–ø—É long

# –ü–µ—Ä–µ—Ç–≤–æ—Ä–µ–Ω–Ω—è —Ç–µ—Å—Ç–æ–≤–∏—Ö –¥–∞–Ω–∏—Ö —É —Ç–µ–Ω–∑–æ—Ä–∏
X_test = torch.tensor(X_test.values).float()  # –ü–µ—Ä–µ—Ç–≤–æ—Ä—é—î–º–æ –Ω–∞ numpy, –∞ –ø–æ—Ç—ñ–º –Ω–∞ —Ç–µ–Ω–∑–æ—Ä —Ç–∏–ø—É float
y_test = torch.tensor(y_test_encoded).long()  # –ü–µ—Ä–µ—Ç–≤–æ—Ä—é—î–º–æ —á–∏—Å–ª–æ–≤—ñ –º—ñ—Ç–∫–∏ –Ω–∞ —Ç–µ–Ω–∑–æ—Ä —Ç–∏–ø—É long
```

```python
print(X_train[:1])
print(y_train[:10])
```

```python
#  –°—Ç–≤–æ—Ä–µ–Ω–Ω—è –º–æ–¥–µ–ª—ñ –Ω–µ–π—Ä–æ–Ω–Ω–æ—ó –º–µ—Ä–µ–∂—ñ
class NObeyesdadNN(nn.Module):
    def __init__(self, input_size, num_clas):
        super(NObeyesdadNN, self).__init__()

        self.fc1 = nn.Linear(input_size, 256)   # –ó–±—ñ–ª—å—à–µ–Ω–æ –∫—ñ–ª—å–∫—ñ—Å—Ç—å –Ω–µ–π—Ä–æ–Ω—ñ–≤ —É –ø–µ—Ä—à–æ–º—É —à–∞—Ä—ñ
        self.bn1 = nn.BatchNorm1d(256)           # Batch Normalization –¥–ª—è 1-–≥–æ —à–∞—Ä—É
        self.fc2 = nn.Linear(256, 128)           # –ó–±—ñ–ª—å—à–µ–Ω–æ –∫—ñ–ª—å–∫—ñ—Å—Ç—å –Ω–µ–π—Ä–æ–Ω—ñ–≤ —É –¥—Ä—É–≥–æ–º—É —à–∞—Ä—ñ
        self.bn2 = nn.BatchNorm1d(128)           # Batch Normalization –¥–ª—è 2-–≥–æ —à–∞—Ä—É
        self.fc3 = nn.Linear(128, num_clas)      # –í–∏—Ö—ñ–¥–Ω–∏–π —à–∞—Ä
        self.relu = nn.ReLU()                    # –ê–∫—Ç–∏–≤–∞—Ü—ñ—è ReLU –¥–ª—è —Å—Ö–æ–≤–∞–Ω–∏—Ö —à–∞—Ä—ñ–≤
        self.dropout = nn.Dropout(0.5)           # Dropout –¥–ª—è —Ä–µ–≥—É–ª—è—Ä–∏–∑–∞—Ü—ñ—ó

    def forward(self, x):
        x = self.relu(self.bn1(self.fc1(x)))  # –ü–µ—Ä—à–∏–π —Å—Ö–æ–≤–∞–Ω–∏–π —à–∞—Ä –∑ Batch Normalization
        x = self.dropout(x)                    # Dropout –ø—ñ—Å–ª—è –ø–µ—Ä—à–æ–≥–æ —à–∞—Ä—É
        x = self.relu(self.bn2(self.fc2(x)))  # –î—Ä—É–≥–∏–π —Å—Ö–æ–≤–∞–Ω–∏–π —à–∞—Ä –∑ Batch Normalization
        x = self.dropout(x)                    # Dropout –ø—ñ—Å–ª—è –¥—Ä—É–≥–æ–≥–æ —à–∞—Ä—É
        x = self.fc3(x)                        # –í–∏—Ö—ñ–¥–Ω–∏–π —à–∞—Ä
        return x
```

```python
model = NObeyesdadNN(X_train.shape[1], 7)  # –ö—ñ–ª—å–∫—ñ—Å—Ç—å –∫–ª–∞—Å—ñ–≤ —è–≤–Ω–æ –∑–∞–¥–∞–Ω–∞ —è–∫ 7

# –í–∏–∑–Ω–∞—á–µ–Ω–Ω—è —Ñ—É–Ω–∫—Ü—ñ—ó –≤—Ç—Ä–∞—Ç —ñ –æ–ø—Ç–∏–º—ñ–∑–∞—Ç–æ—Ä–∞
criterion = nn.CrossEntropyLoss()
optimizer = torch.optim.SGD(model.parameters(), lr=0.01)

num_epoch = 1000

# –°–ø–∏—Å–∫–∏ –¥–ª—è –∑–±–µ—Ä–µ–∂–µ–Ω–Ω—è –≤—Ç—Ä–∞—Ç —Ç–∞ —Ç–æ—á–Ω–æ—Å—Ç—ñ
train_loss = []
test_loss = []
train_accs = []
test_accs = []

# –¢—Ä–µ–Ω—É–≤–∞–Ω–Ω—è –º–æ–¥–µ–ª—ñ
for epoch in range(num_epoch):
    model.train()  # –ü–µ—Ä–µ–º–∏–∫–∞—î–º–æ –º–æ–¥–µ–ª—å –≤ —Ä–µ–∂–∏–º —Ç—Ä–µ–Ω—É–≤–∞–Ω–Ω—è

    # –ü—Ä–æ–≥–Ω–æ–∑
    outputs = model(X_train)

    # –û–±—á–∏—Å–ª–µ–Ω–Ω—è –≤—Ç—Ä–∞—Ç
    loss = criterion(outputs, y_train)
    train_loss.append(loss.cpu().detach().numpy())

    optimizer.zero_grad()  # –û—á–∏—â–∞—î–º–æ –≥—Ä–∞–¥—ñ—î–Ω—Ç–∏
    loss.backward()        # –û–±—á–∏—Å–ª—é—î–º–æ –≥—Ä–∞–¥—ñ—î–Ω—Ç–∏
    optimizer.step()       # –û–Ω–æ–≤–ª—é—î–º–æ –ø–∞—Ä–∞–º–µ—Ç—Ä–∏

    # –¢–æ—á–Ω—ñ—Å—Ç—å –Ω–∞ —Ç—Ä–µ–Ω—É–≤–∞–ª—å–Ω–∏—Ö –¥–∞–Ω–∏—Ö
    _, predicted = torch.max(outputs.data, 1)  # –ó–Ω–∞—Ö–æ–¥–∏–º–æ –∫–ª–∞—Å –∑ –Ω–∞–π–±—ñ–ª—å—à–æ—é –π–º–æ–≤—ñ—Ä–Ω—ñ—Å—Ç—é
    acc = 100 * torch.sum(predicted == y_train).double() / len(y_train)
    train_accs.append(acc)

    if (epoch + 1) % 10 == 0:
        print(f'Epoch [{epoch+1}/{num_epoch}] Loss: {loss.item():.4f}   Acc: {acc.item():.4f}')

    # –¢–µ—Å—Ç—É–≤–∞–Ω–Ω—è –º–æ–¥–µ–ª—ñ
    model.eval()  # –ü–µ—Ä–µ–º–∏–∫–∞—î–º–æ –º–æ–¥–µ–ª—å –≤ —Ä–µ–∂–∏–º —Ç–µ—Å—Ç—É–≤–∞–Ω–Ω—è
    with torch.no_grad():  # –í–∏–º–∏–∫–∞—î–º–æ –æ–±—á–∏—Å–ª–µ–Ω–Ω—è –≥—Ä–∞–¥—ñ—î–Ω—Ç—ñ–≤
        outputs = model(X_test)

        # –û–±—á–∏—Å–ª–µ–Ω–Ω—è –≤—Ç—Ä–∞—Ç
        loss = criterion(outputs, y_test)
        test_loss.append(loss.cpu().detach().numpy())

        # –¢–æ—á–Ω—ñ—Å—Ç—å –Ω–∞ —Ç–µ—Å—Ç–æ–≤–∏—Ö –¥–∞–Ω–∏—Ö
        _, predicted = torch.max(outputs.data, 1)
        acc = 100 * torch.sum(predicted == y_test).double() / len(y_test)
        test_accs.append(acc)
```

```python
plt.figure(figsize=(4, 3))
plt.plot(train_loss, label='Train')
plt.plot(test_loss, label='Validation')
plt.legend(loc='best')
plt.xlabel('Epochs')
plt.ylabel('Cross-Entropy Loss')
plt.title('Training vs Validation Loss')
plt.show()
```

```python
X = df.drop('House_Price', axis=1)
y = df['House_Price']
```

```python
# –ú–∞—Å—à—Ç–∞–±—É–≤–∞–Ω–Ω—è
y_scaler = StandardScaler()
y_scaled = y_scaler.fit_transform(y.values.reshape(-1, 1)).flatten()

X_scaler = StandardScaler()
X_scaled = X_scaler.fit_transform(X)

# –¢—Ä–µ–Ω—É–≤–∞–ª—å–Ω–∏–π —Ç–∞ —Ç–µ—Å—Ç–æ–≤–∏–π –ø–æ–¥—ñ–ª
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y_scaled, test_size=0.2, random_state=42)
```

```python
# –î–∞—Ç–∞—Å–µ—Ç
class HouseDataset(Dataset):
    def __init__(self, X, y):
        self.X = torch.tensor(X, dtype=torch.float32)
        self.y = torch.tensor(y, dtype=torch.float32)

    def __len__(self):
        return len(self.y)

    def __getitem__(self, idx):
        return self.X[idx], self.y[idx]
```

```python
train_dataset = HouseDataset(X_train, y_train)
test_dataset = HouseDataset(X_test, y_test)
```

```python
# Loading Batches of Data

train_dataloader = DataLoader(train_dataset,
                              batch_size=200,
                              num_workers=4
                             )

test_dataloader = DataLoader(test_dataset,
                              batch_size=200,
                              num_workers=4
                             )
```

```python
next(iter(train_dataloader))
```

```python
import torch.nn as nn
class LinearModel(nn.Module):
    def __init__(self, in_dim, out_dim=1):
        super().__init__()

        self.features = nn.Sequential(
            nn.Linear(in_dim, 256),
            nn.BatchNorm1d(256),
            nn.ReLU(),
            nn.Dropout(0.3),

            nn.Linear(256, 128),
            nn.BatchNorm1d(128),
            nn.ReLU(),
            nn.Dropout(0.3),

            nn.Linear(128, 64),
            nn.BatchNorm1d(64),
            nn.ReLU(),
            nn.Dropout(0.3),

            nn.Linear(64, out_dim)
        )

    def forward(self, x):
        return self.features(x)
```

```python
# Initialize the model
model = LinearModel(in_dim=X_train.shape[1], out_dim=1)

# Define the loss function and optimizer
criterion = nn.MSELoss()
optimizer = torch.optim.Adam(model.parameters(), lr=1e-4)

train_losses, train_rmses = [], []
test_losses, test_rmses= [], []
```

```python
num_epochs = 200

# –§—É–Ω–∫—Ü—ñ—è RMSE –Ω–∞ –æ—Å–Ω–æ–≤—ñ MSELoss
def RMSE(y_true, y_pred):
    mse_loss = torch.nn.MSELoss()
    y_true = torch.tensor(y_true, dtype=torch.float32)
    y_pred = torch.tensor(y_pred, dtype=torch.float32)
    mse = mse_loss(y_pred, y_true)
    return torch.sqrt(mse)

for epoch in range(num_epochs):

    # Train step
    model.train()

    y_pred_train = []
    y_true_train = []

    for data in train_dataloader:
        # –û—Ç—Ä–∏–º—É—î–º–æ —Ç–∞ –≥–æ—Ç—É—î–º–æ –¥–∞–Ω—ñ
        inputs, targets = data
        inputs, targets = inputs.float(), targets.float()
        targets = targets.reshape((targets.shape[0], 1))

        # –ü—Ä–æ–≥–Ω–æ–∑ (Forward pass)
        outputs = model(inputs)
        loss = criterion(outputs, targets)

        # –û–Ω–æ–≤–ª–µ–Ω–Ω—è –≤–∞–≥
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

        # –ó–±–µ—Ä—ñ–≥–∞—î–º–æ –ø—Ä–æ–≥–Ω–æ–∑–∏ —ñ —Å–ø—Ä–∞–≤–∂–Ω—ñ –∑–Ω–∞—á–µ–Ω–Ω—è
        y_pred_train.extend(outputs.cpu().detach().numpy())
        y_true_train.extend(targets.cpu().detach().numpy())

    # –û–±—á–∏—Å–ª—é—î–º–æ RMSE –∫–æ–∂–Ω—ñ 10 –µ–ø–æ—Ö
    if (epoch+1) % 10 == 0:
        rmse_train = RMSE(y_true_train, y_pred_train)
        print(f'Epoch [{epoch+1}/{num_epochs}], Loss: {loss.item():.4f}, RMSE: {rmse_train:.4f}')

    train_rmses.append(RMSE(y_true_train, y_pred_train))
    train_losses.append(loss.cpu().detach().numpy())

    # Eval step
    model.eval()

    y_pred_test = []
    y_true_test = []

    with torch.no_grad():
        for data in test_dataloader:
            # –û—Ç—Ä–∏–º—É—î–º–æ —Ç–∞ –≥–æ—Ç—É—î–º–æ –¥–∞–Ω—ñ
            inputs, targets = data
            inputs, targets = inputs.float(), targets.float()
            targets = targets.reshape((targets.shape[0], 1))

            # –ü—Ä–æ–≥–Ω–æ–∑ (Forward pass)
            outputs = model(inputs)
            loss = criterion(outputs, targets)

            # –ó–±–µ—Ä—ñ–≥–∞—î–º–æ –ø—Ä–æ–≥–Ω–æ–∑–∏ —ñ —Å–ø—Ä–∞–≤–∂–Ω—ñ –∑–Ω–∞—á–µ–Ω–Ω—è
            y_pred_test.extend(outputs.cpu().detach().numpy())
            y_true_test.extend(targets.cpu().detach().numpy())

        # –û–±—á–∏—Å–ª—é—î–º–æ RMSE –¥–ª—è —Ç–µ—Å—Ç–æ–≤–∏—Ö –¥–∞–Ω–∏—Ö
        test_rmses.append(RMSE(y_true_test, y_pred_test))
        test_losses.append(loss.cpu().detach().numpy())
```

```python
plt.figure(figsize=(4, 3))
plt.plot(train_losses, label='Train')
plt.plot(test_losses, label='Validation')
plt.legend(loc='best')
plt.xlabel('Epochs')
plt.ylabel('MAE')
plt.title('Training vs Validation Loss')
plt.show()
```
